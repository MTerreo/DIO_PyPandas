```python
#A coluna PRODUÇÃO está como tipo objeto, para por fazer alguns cálculos vou transformar para tipo float64
#passo 1 = passar para tipo string
#passo 2 = substituir a virgula por ponto
#passo 3 = passar para tipo float64

#alterando o tipo de dados da coluna PRODUÇÃO
df["PRODUÇÃO"]=df["PRODUÇÃO"].astype("string")
```

```python
#Somente para verificar o tipo de dado da coluna "PRODUÇÂO" agora é "string"
df.dtypes
```

```python
#Substituindo  a virgula por ponto para permitir a mundança para type float64
df["PRODUÇÃO"] = df["PRODUÇÃO"].str.replace(",", ".")
```

```python
#alterando o tipo de dados da coluna LojaID
df["PRODUÇÃO"] = df["PRODUÇÃO"].astype("float64")
```

```python
#Somente para verificar o tipo de dado da coluna "PRODUÇÂO" agora é "float64"
df.dtypes
```

```python
#agora vou verificar a evolução da produção anual total
#Agrupamento por cidades
df.groupby("ANO")["PRODUÇÃO"].sum()
```

```python
#Para apresentar na forma de gráfico de barras, com titulo e nome dos eixos vou importar o matplotlib
import matplotlib.pyplot as plt
import numpy as np
df.groupby("ANO")["PRODUÇÃO"].sum().plot.bar(title="Produção x Ano", color="green")
plt.xlabel("Ano")
plt.ylabel("Produção");
```

```python
#Passando a coluna "GRANDE REGIÃO" para tipo string
df["GRANDE REGIÃO"]=df["GRANDE REGIÃO"].astype("string")
df["GRANDE REGIÃO"].unique()
```

```python
#Separando por região
df_sul = df[df["GRANDE REGIÃO"] == "REGIÃO SUL"]
df_sudeste = df[df["GRANDE REGIÃO"] == "REGIÃO SUDESTE"]
df_norte = df[df["GRANDE REGIÃO"] == "NORTE"]
df_nordeste = df[df["GRANDE REGIÃO"] == "NORDESTE"]
df_centro = df[df["GRANDE REGIÃO"] == "REGIÃO CENTRO-OESTE"]
```

```python
#Vendo a evolução da produção para uma região
df_sul.groupby("ANO")["PRODUÇÃO"].sum().plot.bar(title="Produção x Ano", color="green")
plt.xlabel("Ano")
plt.ylabel("Produção")
plt.legend();
```

```python
#O mesmo pode ser feito para as outras regiões
df_sudeste.groupby("ANO")["PRODUÇÃO"].sum().plot.bar(title="Produção x Ano", color="darkblue")
plt.xlabel("Ano")
plt.ylabel("Produção");
plt.legend();
```

```python
#Comparando a evolução da Produção em duas regiões
labels = df["ANO"].unique()
sul_subtot = df_sul.groupby("ANO")["PRODUÇÃO"].sum().array
sudeste_subtot = df_sudeste.groupby("ANO")["PRODUÇÃO"].sum().array

x = np.arange(len(labels))  # the label locations
width = 0.45  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, sul_subtot, width, label='Sul', color="green")
rects2 = ax.bar(x + width/2, sudeste_subtot, width, label='Sudeste', color="darkblue")


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Produção')
ax.set_title('Produção por Região')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.xticks(rotation='vertical')  #Faz o ano aparecer rotacionado
fig.tight_layout(pad=0.1, h_pad=200, w_pad=200 , rect=None)

plt.show();
```
